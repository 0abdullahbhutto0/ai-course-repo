```python
from sklearn.neighbors import KNeighborsClassifier
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = {
    'Mineral' : ['Sedimentry','Sedimentry','Sedimentry','Sedimentry','Metamorphic','Metamorphic','Metamorphic','Metamorphic'],
    'Hardness' : [3,2,4,4,2,3,4,5],
    'Density' : [4,3,3,2,5,2,5,3]
}
sample_data = {
    'Hardness' : [2,4],
    'Density' : [1,1]
}
data_df = pd.DataFrame(data)
sampleData_df = pd.DataFrame(sample_data)
print(data_df)
print('\n')
print(sampleData_df)
```

```python
plt.scatter(data_df['Hardness'].values,data_df['Density'].values, marker=() )
plt.scatter(sampleData_df['Hardness'].values,sampleData_df['Density'].values,marker='*',color='red')
plt.ylim(bottom=0)
plt.xlim(left=0)
plt.show()
```

```python
X = data_df[['Hardness', 'Density']].values
y = data_df['Mineral'].values
```

```python
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X,y)
```

```python
pred = knn.predict(sampleData_df[['Hardness', 'Density']].values)
print(pred)
```
